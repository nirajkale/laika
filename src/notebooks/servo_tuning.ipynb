```python
import matplotlib.pyplot as plt
import math
```

```python
target = 180
current_pos = 0
beta1 = 0.75
step = 1 
poins = []

poins.append(current_pos)
while current_pos < target:
    current_pos = math.ceil(current_pos + (target - current_pos) * beta1)
    step = step + 1
    poins.append(current_pos)

plt.plot(poins)
plt.scatter(range(len(poins)), poins)
```
